```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
mpl.style.use('ggplot')
```

```python
car_1 =  pd.read_csv("../../data/good_car_data.csv")
car_1.head(5)
```

```python
#encoding for name, company and fuel type
#1 name
car_name_counts = car_1['name'].value_counts()
car_name_counts
```

```python
name_threshold=10
name_counts_below_replace = car_name_counts[car_name_counts <= name_threshold].index
name_counts_below_replace
```

```python
name_dummiew_car_df = pd.get_dummies(car_1['name'].replace(name_counts_below_replace,"name_unknown"))
name_dummiew_car_df
```

```python
car_name_updated = car_1.join(name_dummiew_car_df, how='inner')
car_name_updated 
```

```python
#car_name_updated=pd.concat([car,name_dummiew_car_df], axis=1 )

car_name_updated=car_name_updated.drop(columns=['name',",".join(car_name_updated.columns[car_name_updated.columns.str.contains('Unnamed', case=False)].tolist())])
```

```python
car_name_updated
```

```python
#company

company_threshold =10

company_counts = car_name_updated["company"].value_counts()
company_counts.index

xx=company_counts.index[company_counts <=10]
xx

```

```python

company_dummy=pd.get_dummies(car_name_updated['company'].replace(xx,'company_unknown'), drop_first=True)
company_dummy
```

```python
car_updated = pd.merge(car_name_updated,company_dummy,left_index=True,right_index=True,how='inner').drop(columns=['company'])
car_updated
```

```python
car_updated['year']
```

```python

#car_year_list_asc=(pd.Series(car_updated['year'].unique()).sort_values(axis=0, ascending=True).astype(str).tolist())
# without converting to string
car_year_list_asc=(pd.Series(car_updated['year'].unique()).sort_values(axis=0, ascending=True).tolist())
car_year_list_asc
```

```python
type(car_year_list_asc)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
X=car_updated.loc[:,car_updated.columns!='Price']
y=car_updated['Price']
```

```python
X
```

```python
X['fuel_type'].unique()
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
transformer= ColumnTransformer(transformers=[
    ('ohe_1', OneHotEncoder(sparse=False, drop='first'),['fuel_type']),
    ('ord_1', OrdinalEncoder(categories=[car_year_list_asc]),['year'])
], remainder='passthrough')
```

```python
transformer
```

```python
transformer.transformers
```

```python
new_x_train=transformer.fit_transform(X_train)
```

```python
new_x_train.shape
```

```python
new_x_train[0]
```

```python
new_x_test= transformer.transform(X_test)
```

```python
new_x_test.shape
```

```python
lr=LinearRegression()
model=lr.fit(new_x_train,y_train)
```

```python
pred=model.predict(new_x_test)
```

```python
r2_score(y_test,pred)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
mean_squared_error(y_test,pred, squared=False)
mean_absolute_error(y_test, pred)
```
